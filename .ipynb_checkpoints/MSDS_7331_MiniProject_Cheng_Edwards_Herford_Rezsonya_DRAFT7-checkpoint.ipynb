# Create Models

## 2. Create The Binary Response Variable 

In this step, we will create the binary response of the college enrollment percentage for the future classification analysis. 

```python
# describe of two response variables :F_Female_ENROLL_sch_pct,M_Male_ENROLL_sch_pct
df_final[['F_Female_ENROLL_sch_pct','M_Male_ENROLL_sch_pct']].describe().transpose()
```

```python
# Unpivots df_final on F_Female_ENROLL_sch_pct and M_Male_ENROLL_sch_pct
# to create a gender_college_enroll for college enrollment model

valuevar = ["F_Female_ENROLL_sch_pct","M_Male_ENROLL_sch_pct"]
idvar = list(df_final)
idvar.remove('F_Female_ENROLL_sch_pct')
idvar.remove('M_Male_ENROLL_sch_pct')

df_operation = pd.melt(df_final, id_vars = idvar,                                            
                        value_vars=valuevar,
                        var_name='gender_college_enroll', value_name='ENROLL_sch_pct')
df_operation.info()
```

In order to equalize class bias, we will split on 50% enrollment percentage. 

```python
#  Binary attributes for collage enrollment of female and male.
#  binary_F_Female_ENROLL_sch_pct,binary_M_Male_ENROLL_sch_pct
#  1: Collage Enrollment Percentage greater than 50%; 
#  0: Collage Enrollment Percentage less than or equal to 50%;

def f(row):
    if row['gender_college_enroll'] == 'F_Female_ENROLL_sch_pct':
        val = 1
    else:
        val = 0
    return val
df_operation['binary_gender_ENROLL_sch'] = df_operation.apply(f, axis=1)


def f(row):
    if row['ENROLL_sch_pct'] > 50:
        val = 1
    else:
        val = 0
    return val
df_operation['binary_ENROLL_sch_pct'] = df_operation.apply(f, axis=1)
```

After we create the binary response, we double check for excessive class bias below. 

```python
#check for class bias based on our thresholds used.  
df_operation['binary_ENROLL_sch_pct'].value_counts()
```

We are seeing a 3-1 class bias. This is concerning, but will run the models using the 50% split for ease of interpretation. We could downsample to match class counts, but in this mini-project we will proceed as is. We will view any simple holdout test results with caution. 

Now, the binary response variable will be assigned to dataframe df_target and the features assigned to dataframe df_feat.

```python
df_target = df_operation.binary_ENROLL_sch_pct
df_feat = df_operation.drop(['binary_ENROLL_sch_pct','ENROLL_sch_pct','gender_college_enroll'], axis=1)
```

### 3.2  Recursive Feature Elimination.

**The Recursive Feature Elimination (RFE)** method is a feature selection approach. It works by recursively removing attributes and building a model on those attributes that remain. It uses the model accuracy to identify which attributes (and combination of attributes) contribute the most to predicting the target attribute.

We use RFE to select 20 explanation attributes for the model.

The following code uses RFE to select attributes.

```python
# Code reference: https://machinelearningmastery.com/feature-selection-in-python-with-scikit-learn/
# Reference: http://scikit-learn.org/stable/modules/feature_selection.html
# Recursive Feature Elimination
# This section might take a few minutes

from sklearn import datasets
from sklearn.linear_model import LogisticRegression

# create a base classifier used to evaluate a subset of attributes
model = LogisticRegression()
# create the RFE model and select attributes count that wish to keep in the model
keep_attributes = 20 
rfe = RFE(model, keep_attributes)
rfe = rfe.fit(df_feat,df_target)
# summarize the selection of the attributes
# print(rfe.support_)
# print(rfe.ranking_)

# Set display parameter
#pd.set_option('display.height', 300)
#pd.set_option('display.max_rows', 300)

# Create dataframe for feature selection output. 
# Selected attributes: Support = true
df_column = list(df_feat)
df_column_selection = pd.DataFrame()
df_column_selection['column']=df_column
df_column_selection['support']=rfe.support_
df_column_selection['ranking']=rfe.ranking_

print(df_column_selection.sort_values('ranking',ascending=True))
```

# Create Models

## 4. Training & Test Dataset Split

We used train_test_split to perform a train and test split (used 80/20 training/testing split). For this mini-project, we will use a simple holdout test set and not perform any iterative k-fold tests. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_feat, df_target, test_size=0.20, random_state=3001)
```

# Create Models

## 5. Normalize The Data

We used the StandardScaler function in scikit-learn to normalize the attributes based on the mean and standard deviation of each column. Doing this will resolve problems if there are any wide range of differences in attributes.

```python
from sklearn.preprocessing import StandardScaler

# we want to normalize the features based upon the mean and standard deviation of each column. 
# However, we do not want to accidentally use the testing data to find out the mean and std (this would be snooping)
# to Make things easier, let's start by just using whatever was last stored in the variables:
##    X_train , y_train , X_test, y_test (they were set in a for loop above)

# scale attributes by the training set
scl_obj = StandardScaler()
scl_obj.fit(X_train) # find scalings for each column that make this zero mean and unit std
# the line of code above only looks at training data to get mean and std and we can use it 
# to transform new feature data

X_train_scaled = scl_obj.transform(X_train) # apply to training
X_test_scaled = scl_obj.transform(X_test) # apply those means and std to the test set (without snooping at the test set values)
```

# Create Models

## 6. Logistic Regression (Default vs. Fine Tuned Parameters)

We first created logistic regression model using default parameters, followed by improvements in accuracy by adding parameter selections. To accomplish the final results, we iteratively tuned the model parameters to improve the accuracy score. 

We are now familiar with the knobs available to sklearn.linear_model.LogisticRegression as defined in the following link:<br> 
http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as mt

d_lr_clf = LogisticRegression() # Default Parameters
d_lr_clf.fit(X_train_scaled,y_train)  # train object
d_lr_y_hat = d_lr_clf.predict(X_test_scaled) # get test set predictions

d_lr_acc = mt.accuracy_score(y_test,d_lr_y_hat)
d_lr_conf = mt.confusion_matrix(y_test,d_lr_y_hat)
print('===========Logistic regression model accuracy (with default parameters)=========')
print('Accuracy:', d_lr_acc )
print('Confusion matrix:')
print(d_lr_conf[0,0], d_lr_conf[0,1])
print(d_lr_conf[1,0], d_lr_conf[1,1])
print('=================================================================================')
print

lr_clf = LogisticRegression(penalty='l1', C=.1, tol=1e-7, intercept_scaling=6, solver='liblinear') # Tuned these parameters
lr_clf.fit(X_train_scaled,y_train)  # train object
lr_y_hat = lr_clf.predict(X_test_scaled) # get test set predictions

lr_acc = mt.accuracy_score(y_test,lr_y_hat)
lr_conf = mt.confusion_matrix(y_test,lr_y_hat)
print('===========Logistic regression model accuracy (with fine tuned parameters)=======')
print('Accuracy:', lr_acc )
print('Confusion matrix:')
print(lr_conf[0,0], lr_conf[0,1])
print(lr_conf[1,0], lr_conf[1,1])
print('=================================================================================')
```

# Create Models

## 7. Support Vector Machines (Default vs. Fine Tuned Parameters)

We first created SVM model using default parameters, followed by improvements in accuracy by adding parameter selections. To accomplish the final results, we iteratively tuned the model parameters to improve the accuracy score. 

We are now familiar with the knobs available to sklearn.svm.SVC as defined in the following link:<br>
http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

```python
from sklearn.svm import SVC

d_svm_clf = SVC() # default parameters
d_svm_clf.fit(X_train_scaled, y_train)  # train object
d_svm_y_hat = d_svm_clf.predict(X_test_scaled) # get test set predictions

d_svm_acc = mt.accuracy_score(y_test,d_svm_y_hat)
d_svm_conf = mt.confusion_matrix(y_test,d_svm_y_hat)
print('===========SVM model accuracy (with default parameters)=========================')
print('Accuracy:', d_svm_acc )
print('Confusion matrix:')
print(d_svm_conf[0,0], d_svm_conf[0,1])
print(d_svm_conf[1,0], d_svm_conf[1,1])
print('=================================================================================')

svm_clf = SVC(C=1.2, kernel='rbf', probability=True, shrinking=True) # tuned these parameters
svm_clf.fit(X_train_scaled, y_train)  # train object
svm_y_hat = svm_clf.predict(X_test_scaled) # get test set predictions

svm_acc = mt.accuracy_score(y_test,svm_y_hat)
svm_conf = mt.confusion_matrix(y_test,svm_y_hat)
print('===========SVM model accuracy (with fine tuned parameters)======================')
print('Accuracy:', svm_acc )
print('Confusion matrix:')
print(svm_conf[0,0], svm_conf[0,1])
print(svm_conf[1,0], svm_conf[1,1])
print('=================================================================================')
```

## 8. Assessment

As shown above, we have accuracies and confusion matrices. This chapter will close out the Create Models section by comparing the accuracies and then providing a few other terms that can be derived from the confusion matrix. 

|     Term    |      Formula     |  Description
|     :-     |     -------------:     |     :-: 
|TN| conf[0,0]|True Negative
|TP| conf[1,1]|True Positive
|FN| conf[1,0]|False Negative
|FP| conf[0,1]|False Positive
|TPR| TP/(TP+FN)  | sensitivity, recall, hit rate, or true positive rate (TPR)
|TNR| TN/(TN+FP)  | specificity or true negative rate (TNR)
|PPV| TP/(TP+FP)  | precision or positive predictive value (PPV)
|NPV| TN/(TN+FN)  | negative predictive value (NPV)
|FNR| FN/(FN+FP)  | miss rate or false negative rate (FNR)
|FPR| FP/(FP+TN)  | fall-out or false positive rate (FPR)
|FDR| FP/(FP+TP)  | false discovery rate (FDR)
|FOR| FN/(FN+TN)  | false omission rate (FOR)
|ACC| (TP+TN)/(TP+TN+FP+FN)  | Accuracy (ACC)
|ERR| (FP+FN)/(TP+TN+FP+FN)  | Error Rate (ERR)
|F1| (2*TP)/((2*TP)+FP+FN)  | F1 score is the harmonic mean of precision and recall. 

Reference:http://www.dataschool.io/simple-guide-to-confusion-matrix-terminology/


The code below provides these terms for our SVM and LR final (fine tuned version) models.

As mentioned in the section outlining our analysis strategy, we decided to conduct a binary classification task of measuring which school features contained in the North Carolina Public High Schools dataset represented the best predictors of whether or not student graduates were likely to enroll in college-level courses using techniques such as recursive feature elimination for preprocessing and modeling our predictions leveraging logistic regression and support vector machine (SVM) techniques. Before receiving model predictions using logistic regression and SVMs, we first had to construct a binary classifier because a binary response variable was not found in the original dataset. Thus, we decided to create two new binary responses. 

The first can be identified as college enrollment percentage (ENROLL_sch_pct), and the second is identified as college enrollment percentage by gender (binary_gender_ENROLL_sch).  After further analysis, we made the decision to focus on college enrollment percentage as our final response variable (ENROLL_sch_pct) for the project. Following identification of a primary binary response variable, we then performed recursive feature elimination to deduce our dataset down to twenty attributes from the original dataset before training our binary classifier. These twenty attributes would be selected using RFE based on their level of contribution for predicting college enrollment of student graduates. In essence, our reasoning for utilizing RFE was due to the high-dimensionality of the original dataset. This technique is often used to remove features of low importance which can improve model accuracy, provide a reduction in model complexity, and decrease risks of overfitting. Furthermore, once we decide to train our models using logistic regression and support vector machines, overall performance with regard to training time will be reduced. 

With regard to model advantages as well as disadvantages using classification techniques such as logistic regression and support vector machines, we did see some observable differences related to model accuracy. Before commenting on observed differences, it’s important to point out the inherent advantages and disadvantages of each technique. Although different from linear regression, logistic regression models are still “linear” in that they function well when classes are linearly separable; however, logistic regression can still be used when classes cannot be easily separated by single decision surface, but the costs from doing so results in coefficients being penalized. 

Some strengths (not exhaustive) of logistic regression are that model outputs will have a good probabilistic interpretation as well as the ability to normalize in order to avoid the pitfall of overfitting. Lastly, models using logistic regression can easily be appended with new data using functions such as stochastic gradient descent or SGD. In terms of weaknesses or disadvantages though, it’s not uncommon to observe underperforming models when classification boundaries are non-linear and/or contain multiple decision boundaries. The “linear” component to logistic regression does not contain the flexibility to naturally capture relationships that are higher in complexity.

In the case of support vector machines, the primary benefit comes from kernels that are non-linear in order to model non-linear classification boundaries. Kernels are a mechanism used by support vector machines to measure distance between two observations.  The goal is to maximize or find the optimal distance between the closest members which each belong to different classes. Unlike logistic regression, support vector machines provide more flexibility when classifying non-linear decision boundaries. With several kernels to choose from, SVMs are robust when it comes to risks of overfitting. The robustness of SVMs really begins to show when modeling datasets with high dimensionality. Some disadvantages are the amount of memory used in processing, tuning models become more difficult due to the number of kernels present, and scalability becomes problematic when using on larger datasets.

As stated above, we did observe and experience minor differences between the models using logistic regression and support vector machines.

# Interpret Support Vectors
Look at the chosen support vectors for the classification task. Do these provide any insight into the data? Explain.

A Support Vector Machine (SVM) is a discriminative classifier formally defined by a separating hyperplane. In other words, given labeled training data (supervised learning), the algorithm outputs an optimal hyperplane which categorizes new examples. In two dimentional space this hyperplane is a line dividing a plane in two parts where in each class lay in either side.

In simple words: **The ultimate goal of SVMs is the separation of classes.**

Reference:https://medium.com/machine-learning-101/chapter-2-svm-support-vector-machine-theory-f0812effc72

To see if these provide any insight into the data, let's look at some of the sklearn.svm.SVC attributes from our fine tuned final SVM model.
