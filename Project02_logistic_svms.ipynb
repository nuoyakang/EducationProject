#### Load Data & Organize Data for Final Analysis

Load required libraries and set the working directory.

```python
# CODE REFERENCE: https://github.com/jakemdrew/EducationDataNC
# Alteration from reference code include removal of schoolYear as these will be fixed years in the following section. 

#import required Libraries
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.model_selection import ShuffleSplit
from sklearn.feature_selection import RFE

from pandas.tools.plotting import boxplot

#**********************************************************************************
# Set the following variables before running this code!!!
#**********************************************************************************

#Location where copies of the raw data files will be downloaded and saved as csv files.
directory = 'C:/Conda/STUDY/data_mining/term_project/project1/HighML/'
schDir = 'C:/Conda/STUDY/data_mining/term_project/project1/HighSch/'
```

Read each data set and combine them as the final data set.

```python
#### Reference: Snippets of this code block was taken directly from Eric Larson git repository: https://github.com/eclarson ####
#### Reference: https://stackoverflow.com/questions/12555323/adding-new-column-to-existing-dataframe-in-python-pandas

df_2014ML = pd.read_csv(directory + 'PublicHighSchools2014_ML.csv') # read in the csv file
df_2014ML['year'] = 2014
print('*********************************Before: 2014 Public High School Data*********************************')
df_2014ML.info(verbose=False)

df_2015ML = pd.read_csv(directory + 'PublicHighSchools2015_ML.csv') # read in the csv file 
df_2015ML['year'] = 2015
print('*********************************Before: 2015 Public High School Data*********************************')
df_2015ML.info(verbose=False)

df_2016ML = pd.read_csv(directory + 'PublicHighSchools2016_ML.csv') # read in the csv file 
df_2016ML['year'] = 2016
print('*********************************Before: 2016 Public High School Data*********************************')
df_2016ML.info(verbose=False)

df_2017ML = pd.read_csv(directory + 'PublicHighSchools2017_ML.csv') # read in the csv file 
df_2017ML['year'] = 2017
print('*********************************Before: 2017 Public High School Data*********************************')
df_2017ML.info(verbose=False)
```

```python
#Combine data
#### Reference https://pandas.pydata.org/pandas-docs/stable/merging.html
df_final = pd.concat([df_2014ML,df_2015ML,df_2016ML,df_2017ML], axis=0, join='inner', ignore_index=True)
print('*********************************After: Combined Public High School Data*********************************')
print('*********************************Verbose=True to show all column names and type of data********************')
df_final.info(verbose=True)
```

A brief data summary.

```python
pd.set_option('display.max_rows', 300)
df_final.describe(include='all').transpose()
```

```python
df_final['ENROLL_sch_pct'] = df_final['F_Female_ENROLL_sch_pct'] + df_final['M_Male_ENROLL_sch_pct']
```

Set the binary response variable. I think the SAT score will be a decent representor of whether a student is going to college or not and a student is taking AP is also representing the desire to go to college. Therefore, I will set both AP and SAT score in the conditional statement to create the binary response.

```python
# Code REFERENCE: https://stackoverflow.com/questions/21702342/creating-a-new-column-based-on-if-elif-else-condition

def f(df):
    if df['sat_avg_score_num']>1100 and df['ap_pct_3_or_above'] >0.34 and df['ENROLL_sch_pct']>0.5:
        val =1
    else:
        val =0
    return val
# We are certain there's a better way to do this, but other options gave us various issues....

df_final['response_class'] = df_final.apply(f, axis=1)
```

```python
#check for class bias based on our thresholds used.  
df_final['response_class'].value_counts()
```

```python
df_final.drop(['unit_code','year'],axis=1).groupby('response_class').mean().transpose()
```

```python
df_final.info()
```

```python
X = df_final.drop(['response_class','unit_code'],axis=1)
y = df_final['response_class']
```

#### Logistic Regression Feature Selection

```python
# create an instance of logistic regression
logistic = LogisticRegression() 
X = df_final.drop(['response_class','unit_code'],axis=1)
y = df_final['response_class']

# create the RFE model and select attributes number that wish to keep in the model
keep_attributes = 20 
rfe = RFE(logistic, keep_attributes)
rfe = rfe.fit(X,y)

# Create dataframe for feature selection output
df_column = list(X)
df_column_selection = pd.DataFrame()
df_column_selection['column']=df_column

# The mask of selected features.
df_column_selection['support']=rfe.support_

# The feature ranking, such that ranking_[i] corresponds to the ranking position of the i-th feature. 
df_column_selection['ranking']=rfe.ranking_

# Create dataset with selected attributes for building logistic regression model
selected_columns = df_column_selection.loc[df_column_selection['support'] == True]['column'].tolist()

print(pd.DataFrame({'selected column':selected_columns}))

print(df_column_selection.sort_values('ranking',ascending=True))
X = X[selected_columns]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.20, random_state=2018)
# scale attributes by the training set
scl_obj = StandardScaler()
scl_obj.fit(X_train) # find scalings for each column that make this zero mean and unit std
# the line of code above only looks at training data to get mean and std and we can use it 
# to transform new feature data
X_train_scaled = scl_obj.transform(X_train) # apply to training
X_test_scaled = scl_obj.transform(X_test) # apply those means and std to the test set (without snooping at the test set values)

# fit the model 
logistic.fit(X_train_scaled,y_train)  # train object

#get the predictions
predictions = logistic.predict(X_test_scaled) # get test set precitions
```

```python

svm_model = SVC(C=0.5, kernel='rbf', degree=3, gamma='auto')
svm = svm_model.fit(X_train_scaled,y_train)

from sklearn.model_selection import ShuffleSplit

num_cv_iterations = 10
num_instances = len(y)
cv_object = ShuffleSplit(n_splits=num_cv_iterations,
                         test_size  = 0.2)

for train_indices, test_indices in cv_object.split(X.values,y.values): 

    X_train = X.values[train_indices]
    y_train = y.values[train_indices]
    
    X_test = X.values[test_indices]
    y_test = y.values[test_indices]
#input value will be the last value returned  
X_train_scaled = scl_obj.transform(X_train) # apply to training
X_test_scaled = scl_obj.transform(X_test)

predict_svc = svm_model.predict(X_test_scaled)
# now let's get the accuracy and confusion matrix
accuracy = accuracy_score(y_test,predict_svc)
conf_matrix = confusion_matrix(y_test,predict_svc)
print("========================================================")
print('The accuracy score is:',accuracy_score(y_test, predict_svc, normalize=True))
print('========================================================')
print('The confusion matrix is:\n', confusion_matrix(y_test,predict_svc))
print('========================================================')    
print('The classification report is:\n'+classification_report(y_test,predict_svc))
```
